```python
def frecuencias_cat(df_cat):
    resultado = df_cat.apply(lambda x: x.value_counts(normalize = True)).T.stack()\
                .to_frame().reset_index()\
                .rename(columns={'level_0':'Variable','level_1':'Valor',0: "Frecuencia"})\
                .sort_values(by = ['Variable','Frecuencia'])
    return(resultado)
```

```python
pd.set_option('display.max_rows', None)

frecuencias_cat(cat[incluir])
```

### Análisis gráfico

```python
def graficos_eda_categoricos(cat):
    
    # Calculamos el número de filas que necesitamos
    from math import ceil
    filas = ceil(cat.shape[1] / 2)

    # Definimos el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    # Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat 

    # Creamos el bucle que va añadiendo gráficos
    for cada, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

```python
graficos_eda_categoricos(cat[incluir])
```

## EDA NUMÉRICAS

### Análisis estadístico

```python
def estadisticos_cont(num):
    #Calculamos describe
    estadisticos = num.describe().T
    #Añadimos la mediana
    estadisticos['median'] = num.median()
    #Reordenamos para que la mediana esté al lado de la media
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    #Lo devolvemos
    return(estadisticos)
```

```python
estadisticos_cont(num)
```

### Análisis gráfico

En este caso el análisis gráfico que tiene más sentido es el de ventas y precios por diferentes dimensiones.

Para ello se volverán a unir los dos dataframes.

```python
df = pd.concat([cat,num], axis=1)

df.head()
```

#### Tendencia global de las ventas

```python
df.groupby('date').ventas.sum().plot();
```

Puede verse como por alguna razón en Septiembre de 2013 se dispararon las ventas. En una situación real se contactaría con un responsable de negocio para tratar de ver el motivo y se incorporaría alguna variable indicador que estaría a 1 durante todos esos días, para que el modelo lo tuviese en cuenta. También vemos ese efecto en Enero, con 0 ventas, deberíamos comprobar si eso está recogido en la variable calendario, y si no incluir otro indicador.

#### Tendencia de las ventas por producto

```python
df.groupby(['date','item_id']).ventas.sum().unstack().plot(subplots = True,  layout = (5,2), sharex=False, figsize=(20,30));
```

Conclusiones:
